# Kmeans without PCA 

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

K_val=range(2,15)
silhouette_score_list=[]

for i in K_val:
    kmeans=KMeans(n_clusters=i,random_state=42)
    label=kmeans.fit_predict(df)
    score=round(silhouette_score(df,label),2)
    silhouette_score_list.append(score)
    
```

```python
silhouette_score_list
```

```python
px.line(x= K_val, y= silhouette_score_list, markers= True, labels= {'x' : 'No of clusters', 'y' : 'Silhoute Scores'})
```

# kmeans with PCA

```python
df_pca=df.drop('clusters',axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

lc=LabelEncoder()
df_pca['city']=lc.fit_transform(df_pca['city'])

df_pca['city']
```

```python
scaling_cols = df.drop('city', axis= 1).columns
scaling_cols
```

```python
from sklearn.preprocessing import MinMaxScaler

mx = MinMaxScaler()

df[scaling_cols] = mx.fit_transform(df[scaling_cols])

df
```

```python
from sklearn.decomposition import PCA

pca=PCA(n_components=2,random_state=42)
df_pca=pca.fit_transform(df_pca)
df_pca
```

```python
df_pca=pd.DataFrame(df_pca,columns=['pca_1','pca_2'])
df_pca
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

k_values = range(2, 15)
silhouette_score_list = []

for k in k_values:

    kmeans = KMeans(n_clusters= k, random_state= 42)

    labels = kmeans.fit_predict(df_pca)

    score = round(silhouette_score(df_pca, labels), 2)

    silhouette_score_list.append(score)
```

```python
silhouette_score_list
```

```python
px.line(x= k_values, y= silhouette_score_list, markers= True, labels= {'x' : 'No of clusters', 'y' : 'Silhoute Scores'})
```

```python
kmeans = KMeans(n_clusters= 3, random_state= 42)
 
df_pca['clusters_pca'] = kmeans.fit_predict(df_pca)

df_pca
```
